```python
import requests

car_data_url = "https://raw.githubusercontent.com/SophieMwangi29/LuxDevHQ-Assignments/refs/heads/main/Kenyan%20Vehicle%20Data.json"

raw_data = requests.get(car_data_url)

raw_data.status_code


data = raw_data.json()
```

```python
import pandas

len(data)

subset = data[:100]


## creating CSV

### 1. convert it into dataframe - it was json

subset_df = pandas.DataFrame(subset)

subset_df.to_csv("subset_df")

subset_df
```
