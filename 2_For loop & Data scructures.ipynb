Question 9: An API provider gives 50,000 free requests. A company makes 350,000 requests. Each extra 1,000 requests costs $0.40. Calculate total cost in USD and KSh at an exchange rate of 129.

```python
free_requests = 50000
tot_requests = 350000
extra_requests_bundle = 1000
extra_requests_bundle_costs = 0.40
exch_rate = 129


non_free_requests = tot_requests - free_requests

cost_of_non_free_requests_USD = ((non_free_requests / extra_requests_bundle) * extra_requests_bundle_costs)
cost_of_non_free_requests_LCY = (extra_requests_bundle_costs * non_free_requests)/extra_requests_bundle * exch_rate

print(cost_of_non_free_requests_USD)
print(cost_of_non_free_requests_LCY)
```
